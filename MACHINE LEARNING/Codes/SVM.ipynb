```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[4, 1], [4, -1], [6, 0], [1, 0], [0, 1], [0, -1]])
y = np.array([1, 1, 1, -1, -1, -1])
```

```python
def manual_svm(x, y, learning_rate=0.01, epochs=100):
    w = np.random.randn(2)
    b = np.random.randn()

    for _ in range(epochs):
        for i, xi in enumerate(x):
            if y[i] * (np.dot(xi, w) + b) < 1:
                w -= learning_rate * (-y[i]*xi)
                b -= learning_rate * (-y[i])
            else:
                w -= learning_rate * 0.01 * w      # small regularization term

    return w, b
```

```python
def predict(x_new, w, b):
    decision = np.dot(x_new, w) + b
    return np.sign(decision)
```

```python
w, b = manual_svm(x, y)

plt.scatter(x[:, 0], x[:, 1], c=y, s=100, cmap=plt.cm.Paired, edgecolor='k')

x1 = np.linspace(-1, 7, 100)
x2 = (-b -w[0] * x1) / w[1]
plt.plot(x1, x2, 'r--')
plt.title("Hyperplane with Manual svm")
plt.xlabel("x1")
plt.ylabel("x2")
plt.grid()
plt.show()
```
