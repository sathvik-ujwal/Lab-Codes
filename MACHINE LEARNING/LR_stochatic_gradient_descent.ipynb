```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(
    {
        "x" : [1,2,4,3,5],
        "y" : [1,3,3,2,5]
    }
)

x = df["x"]
y = df["y"]
x = x.to_numpy()
y = y.to_numpy()

n = len(x)
df
```

```python
# LINEAR REGRESSION USING STOCHASTIC GRADIENT DESCENT
# FIRST INITIALIZE BOTH COEFFICENTS TO 0.0
B0 = 0
B1 = 0

# alpha is the learning rate
alpha = 0.015

for _ in range(4):
    for j in range(n):
        p = B0 + B1*x[j]
        error = p - y[j]
        B0 = B0 - alpha * error
        B1 = B1 - alpha * error * x[j]
        
print(f"Final value of intercept B0 : {B0}")
print(f"Final value of slope B1 : {B1}")
```

```python
# MAKING PREDICTIONS AND EVALUATION
from sklearn.metrics import mean_squared_error

y_pred = B0 + B1*x
RMSE = mean_squared_error(y,y_pred, squared = False)
print(f"Root mean squared error : {RMSE}")
```

```python
plt.scatter(x,y, label="Data points")
plt.plot(x, y_pred,label = "Regression line")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear regression using stochastic gradient descent")
plt.legend()
plt.show()
```
